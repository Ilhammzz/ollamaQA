````python
query = "Jelaskan seluruh isi dari konteks yang diberikan!"
````

````python
%%time
resp = llm.complete(query_template.format(query))
display(Markdown(resp.text))
````

````python
%%time
resp = llm.stream_complete(query_template.format(query))
for r in resp:
    print(r.delta, end="")
print("\n")  # space for %%time
````
